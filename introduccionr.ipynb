## (h) Create a 2-dimensional array of 3 rows and 4 columns with random integer values. Then, calculate the sum of each row and column.

```r
# Crea una matriz bidimensional de 3 filas y 4 columnas con valores aleatorios
matriz <- matrix(sample(1:10, 12, replace = TRUE), nrow = 3, ncol = 4)

# Calcula la suma de cada fila
sumas_filas <- rowSums(matriz)

# Calcula la suma de cada columna
sumas_columnas <- colSums(matriz)

# Imprime la matriz y las sumas
cat("Matriz:\n")
print(matriz)
cat("Sumas de filas:", sumas_filas, "\n")
cat("Sumas de columnas:", sumas_columnas, "\n")
```

## (k) Load a CSV file into a DataFrame and then calculate the mean, median, and mode of one of the columns in the DataFrame

```r
# Cargar el archivo CSV en un DataFrame
dataframe <- read.csv('/kaggle/input/bengaluruhousedata12/BengaluruHouseData.csv')

# Eliminar filas con valores NA
dataframe <- na.omit(dataframe)

# Seleccionar la columna para los cálculos
columna <- 'price'

# Calcular la media
media <- mean(dataframe[[columna]])

# Calcular la mediana
mediana <- median(dataframe[[columna]])

# Calcular la moda
moda <- as.numeric(names(which.max(table(dataframe[[columna]]))))

# Imprimir los resultados
cat("Media:", media, "\n")
cat("Mediana:", mediana, "\n")
cat("Moda:", moda, "\n")
```

```r
data <- data[complete.cases(data), ]
#data
data$Ozone <- as.integer(data$Ozone)
data$Solar.R <- as.integer(data$Solar.R)
data$Solar.R <- as.integer(data$Solar.R)


media_Ozone <- mean(x=data$Ozone, na.rm=TRUE)
media_Solar.R <- mean(x=data$Solar.R, na.rm=TRUE)


data$Ozone[is.na(data$Ozone)] <- media_Ozone
data$Solar.R[is.na(data$Solar.R)] <- media_Solar.R
data
```

## (w) For each month, obtain the minimum and the maximum temperature registered in Celsius.


```r
temperature_summary <- data %>%
  group_by(Month) %>%
  summarize(Temperature_Min = min(TempC), Temperature_Max = max(TempC))

temperature_summary
```

## (x) Obtain the average of the following parameters by month: temperature in celsius, wind, solar radiation and ozone.


```r
parameter_summary <- data %>%
  group_by(Month) %>%
  summarize(Average_Temperature = mean(TempC),
            Average_Wind = mean(Wind),
            Average_Solar_Radiation = mean(Solar.R),
            Average_Ozone = mean(Ozone))

parameter_summary
```
